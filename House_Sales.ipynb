```python
# 2.1
# Import libraries
import pandas as pd
# Import dataset
myDf = pd.read_csv("LOGITData.csv")
myDf
# Find rank of each case based on "SalePrice"
myDf['SalePriceRank'] = myDf['SalePrice'].rank(method = 'first')
myDf.sort_values(by = ['SalePriceRank'])
```

```python
# 2.2
# Make bins
myDf['SalePriceBin'] = pd.cut(myDf['SalePriceRank'], 25)

# Display counts
myDf['SalePriceBinCount'] = pd.Series(myDf['SalePriceBin'].value_counts())
myDf.sort_values(by = ['SalePriceRank'])
```

```python
# 2.3
import numpy as np
import pandas as pd
import seaborn as sns

# Added SalePriceBinNumber for calculation of logit
myDf['SalePriceBinNumber'] =  pd.qcut(myDf['SalePriceRank'], 25, labels=False) + 1

# Sanity check bins, use it to store logit results, and set up for plots
bins = myDf[["SalePriceBin", "SalePriceBinNumber"]].drop_duplicates()
bins.sort_values(by = ['SalePriceBinNumber'])

# Iterate over the 25 bins
for i in range(25):
    # For each bin, create a new small dataframe 
    subdf = myDf[(myDf['SalePriceBinNumber'] == i + 1)]
    # Count number of cases with at least one Fireplace in the bin
    num_cases_w_fps = sum(subdf.Fireplaces > 0)
    # Count total number of cases in the bin
    num_total_cases = subdf.shape[0]
    # Calulate p
    p = (num_cases_w_fps + 0.5) / num_total_cases
    # Calculate logit
    logit = np.log(p / (1 - p))
    # Append logit to results dataframe
    bins.loc[i,"logit"]=logit
    # Calculate the average sale price per bin
    avgsaleprice = subdf["SalePrice"].mean()
    # Add average to the bins dataset
    bins.loc[i,"avgsaleprice"] = avgsaleprice


bins
```
